```python
# read all the files in a folder

import glob
import os

def get_files(folder):
    files =[]
    print(folder)
    for f in os.listdir(folder):
        if os.path.isfile(os.path.join(folder, f)):
            files.append(os.path.join(folder, f))
        elif os.path.isdir(os.path.join(folder, f)):
            newFiles = get_files(os.path.join(folder, f))
            files.extend(newFiles)
    
    return files
```

```python

class Citation:
    def __init__(self, id, author, title, journal, volume, year, pages, isbn, doi, url, abstract, keywords) -> None:
        self.id = id
        self.author = author
        self.title = title
        self.journal = journal
        self.volume = volume
        self.year = year
        self.pages = pages
        self.isbn = isbn 
        self.doi = doi
        self.url = url
        self.abstract = abstract
        self.keywords = keywords
    
    def numbering(self):
        return f'{self.id}. {self.author}, {self.title}, {self.journal}, {self.volume}, {self.year}, {self.pages}, {self.isbn}, {self.doi}, {self.url}.'
    
    def apaCite(self):  
        authors = self.author.split(',')
        # print(authors)

        apaAuthors=''
        k = len(authors)
        for i in range(k):
            a = authors[i]
            a = a.replace('.', '. ')
            names = a.split()
            # print(a, names)
            surname = names[-1]
            others = [n[0] for n in names[:-1]]
            initials = '. '.join(others) + '.'
            
            if k > 2:
                apaAuthors = surname + ' et al.'
                break
            else:
                apaAuthors = apaAuthors + ' and ' + surname
                
        if k < 3:
            apaAuthors = apaAuthors[5:]     
               
        return f'{apaAuthors} ({self.year})'
        
    def apa(self):        
        # author = citation[0].author
        # author = 'Single Author'
        # print(author)

        authors = self.author.split(',')
        # print(authors)

        apaAuthors=''
        k = len(authors)
        for i in range(k):
            a = authors[i]
            a = a.replace('.', '. ')
            names = a.split()
            # print(a, names)
            surname = names[-1]
            others = [n[0] for n in names[:-1]]
            initials = '. '.join(others) + '.'
            
            if k > 1 and i == k-1:
                apaAuthors = apaAuthors + ', & ' + (surname +', ' + initials)
            else:
                apaAuthors = apaAuthors + ', ' + (surname +', ' + initials)

        apaAuthors = apaAuthors[2:]
        volume = self.volume.replace('Volume ','')
        pages = self.pages.replace('Pages ', '')
                                  
        return f'{apaAuthors} ({self.year}). {self.title}. {self.journal}, {volume}, {pages}. {self.doi}. [{self.id}]'
        
    def __repr__(self) -> str:
        return f"Citation(id={self.id}, author={self.author}, title={self.title}, journal={self.journal}, volume={self.volume}, year={self.year}, pages={self.pages}, isbn={self.isbn}, doi={self.doi}, url={self.url}, abstract={self.abstract}, keywords={self.keywords})"
    
```
